```python
import cv2
import pandas as pd
import numpy as np
from scipy.spatial import distance
from google.colab.patches import cv2_imshow
import matplotlib.image as pltimg
import matplotlib.pyplot as plt
from sklearn.preprocessing import RobustScaler, StandardScaler, MinMaxScaler
```

```python
# Detect edge of circle
img = cv2.imread('/content/drive/MyDrive/Machine Learning/Zero Shot Learning/NTECH/Image Data/Shapes/circle.png')
img = cv2.resize(img, (64, 64))
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
edges_circle = cv2.Canny(image=img_gray, threshold1=100, threshold2=200)

# Detect edge of rectangle / square
img = cv2.imread('/content/drive/MyDrive/Machine Learning/Zero Shot Learning/NTECH/Image Data/Shapes/square.png')
img = cv2.resize(img, (64, 64))
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
edges_square = cv2.Canny(image=img_gray, threshold1=100, threshold2=200)

# Detect edge of sample image and compute minkowski distance of image to circle and rectangle
kaua_dist_circle = []  # Store distances from circle to image
kaua_dist_rect = []    # Store distances from circle to rectangle
for i in range(1,11):
  img = cv2.imread('/content/drive/MyDrive/Machine Learning/Zero Shot Learning/NTECH/Image Data/Apple/Apple{}.png'.format(i))
  img = cv2.resize(img, (64, 64))
  img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
  edges_apple = cv2.Canny(image=img_gray, threshold1=100, threshold2=200)
  
  kaua_dist_circle.append(distance.minkowski(edges_circle.flatten(), edges_apple.flatten(), 2))
  kaua_dist_rect.append(distance.minkowski(edges_square.flatten(), edges_apple.flatten(), 2))
```
