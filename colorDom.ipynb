```python
import PIL
from PIL import Image
from matplotlib import pyplot as plt
from scipy.spatial import distance
import numpy as np
```

```python
apple_means = []
for i in range(1,11):
    im = Image.open('/content/drive/MyDrive/Machine Learning/Zero Shot Learning/NTECH/Image Data/Ora/Apple{}.png'.format(i))
    w, h = im.size  
    colors = im.getcolors(w*h)
    mean = [0, 0, 0]
    counter = 0

    result = sorted(colors, key=lambda tup: tup[0], reverse=True)
    
    for res in result:
        if res[1][-1] >200:
            #red = res[0][0]
            mean[0] += res[1][0]
            #green = res[0][1]
            mean[1] += res[1][1]
            #blue = res[0][2]
            mean[2] += res[1][2]
            counter +=1
    apple_means.append([x/counter for x in mean])     
    
```

```python
euc_dist_r = []     # Euclidean Distance of Sample to Red Color
euc_dist_g = []     # Euclidean Distance of Sample to Green Color
euc_dist_b = []     # Euclidean Distance of Sample to Blue Color
euc_dist_o = []     # Euclidean Distance of Sample to Orange Color
euc_dist_bl = []    # Euclidean Distance of Sample to Black Color
for datapoints in apple_means:
  xs = datapoints[0]
  ys = datapoints[1] 
  zs = datapoints[2]
  euc_dist_r.append(distance.euclidean([xs,ys,zs], [255, 0, 0]))
  euc_dist_g.append(distance.euclidean([xs,ys,zs], [0, 255, 0]))
  euc_dist_b.append(distance.euclidean([xs,ys,zs], [0, 0, 255]))
  euc_dist_o.append(distance.euclidean([xs,ys,zs], [255, 165, 0]))
  euc_dist_bl.append(distance.euclidean([xs,ys,zs], [0, 0, 0]))
```

```python
# Check mean of all the euclidean distances
np.array(euc_dist_r).mean(), np.array(euc_dist_g).mean() ,np.array(euc_dist_b).mean(),np.array(euc_dist_o).mean(),np.array(euc_dist_bl).mean()
```
